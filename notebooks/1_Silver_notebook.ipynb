# **Data Reading**

```python
df = spark.read.parquet("Files/bronze")
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df=df.withColumn('Revenue_perUnit',col('Revenue')/col('Units_sold'))
display(df)
```

```python
df=df.withColumn('model_category',split(col('Model_ID'),'-')[0])
display(df)
```

# **AD HOC**

```python
display(df.groupBy('Year','BranchName').agg(sum('Units_sold').alias('Total_units')).sort('Year','Total_units',ascending=[1,0]))
```

# **Data Writing**

```python
df= df.dropDuplicates([
    "Branch_ID", "Dealer_ID", "Model_ID", "Date_ID"
])
```

```python
df.write.format('parquet')\
    .mode('overwrite')\
    .save('abfss://incre_ws@onelake.dfs.fabric.microsoft.com/lake_san.Lakehouse/Files/silver/carsales')\
    
```
